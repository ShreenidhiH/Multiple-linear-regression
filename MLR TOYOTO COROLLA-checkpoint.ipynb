```python
# Checking for Outliers
def outliers(df):
    for i in df:
        plt.figure(figsize=(6,5))
        sn.boxplot(data[i])
        plt.xlabel(i)
        plt.show()
outliers(data.iloc[:,[1,2,3,4,7,8]])
```

#### Almost Every column has Outliers

Apart from Quarterly tax and cc every columns outlier values atre not very high compared to 75% values.

Let's treat Quarterly tax and cc

```python
# Treating outliers
def outliers_treat(df,columns):
    for i in columns:
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        iqr=q3-q1
        low=q1-1.5*(iqr)
        high=q3+1.5*(iqr)
        df[i]=np.where(df[i]<low,low,np.where(df[i]>high,high,df[i]))
    
```

```python
data_1=data.copy()
outliers_treat(data,["Quarterly_Tax","cc"])
```

```python
outliers(data_1.iloc[:,[4,7]])
```

#### Both columns are treated

## Model Building and Subset Selection

```python
# Building model
model1=smf.ols('Price~age+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_1).fit()
model1.summary()
```

### Since p value is greater for cc and Doors , calculate VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.graphics.regressionplots import influence_plot
```

```python
# Separating independent and dependent variable
x=data_c3[['age','KM','HP','cc','Gears','Doors','Quarterly_Tax','Weight']]
y=data_c3["Price"]
```

## 1. Without Any Transformation

```python
model4=smf.ols('Price~age+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=data_c3).fit()
model4.summary()
```

#### Accuracy is 88.8%

## 2.With log transformation to Y

```python
log_y=np.log(y)
log_x=np.log(x)
```

```python
model5=smf.ols('log_y~x',data=data_c3).fit()
model5.summary()
```

#### Accuracy is 85.8%

## 3.With log Transformation to x

```python
model6=smf.ols('y~log_x',data=data_c3).fit()
model6.summary()
```

#### Accuracy is 84.9%

## 4.With Squareroot transformation to x

```python
sq_x=np.sqrt(x)
sq_y=np.sqrt(y)
```

```python
model7=smf.ols('y~sq_x',data=data_c3).fit()
model7.summary()
```

#### Accuracy is 89.2%

## 5.With Squareroot transformation to y

```python
model8=smf.ols('sq_y~x',data=data_c3).fit()
model8.summary()
```

#### Accuracy is 88%

## 6.With Cuberoot transformation  to y

```python
cb_x=np.cbrt(x)
cb_y=np.cbrt(y)
```

```python
model9=smf.ols('cb_y~x',data=data_c3).fit()
model9.summary()
```

#### Accuracy is 87.4%

## 7.With Cuberoot transformation  to x

```python
model10=smf.ols('y~cb_x',data=data_c3).fit()
model10.summary()
```

#### Accuracy is 88.6%

```python
performance_matrix=pd.DataFrame()
```

```python
sqr=[]
adj_r=[]
models=[model4,model5,model6,model7,model8,model9,model10]
for i in models:
    r=np.round(i.rsquared,4)*100
    a_r=np.round(i.rsquared_adj,4)*100
    sqr.append(r)
    adj_r.append(a_r)
sqr    
```

```python
model=["model4","model5","model6","model7","model8","model9","model10"]
transformations=["no Transformation","Log to y","Log to x","Square root to x","Square root to y","Cuberoot to y","Cuberoot to x"]
performance_matrix["models"]=model
performance_matrix["r_squared(in%)"]=sqr
performance_matrix["Adj_R_squared(in%)"]=adj_r
performance_matrix["Transformations"]=transformations
performance_matrix
```

```python
print("Highest accuracy of model is:",max(adj_r))
```

### The model with Squareroot Transformation to X has maximun accuracy.
